```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit_transform(X,y)
```

```python
scaled
```

```python
## Q No. 2 :
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.70, test_size=0.30, random_state=42)
```

```python
## Q no. 3 :
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=100,max_depth=10,min_samples_split=3,random_state=42)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
## Q no. 4 :
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_pred,y_test))
```
